```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
%matplotlib inline 
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('melb_data.csv')
df
```

```python
df.head()
```

```python
df['Price'].plot(kind='hist', figsize=(8, 5), bins = 100)

plt.title(('Distribusi Harga Rumah'), fontsize=15) # add a title to the histogram
plt.ylabel('Count') # add y-label
plt.xlabel('Price') # add x-label

plt.show()
```

```python
df.Distance.plot(kind='box', figsize=(10, 6))

plt.title('Box plot of Japanese Immigrants from 1980 - 2013')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
df_jarak = df['Distance']
df_jarak

df_kota = df['Address']
df_kota
```

```python
# horizontal box plots
#df_jarak.plot(kind='box', figsize=(8, 7), color='blue', vert=False)
sns.boxplot(x=df['Distance'], y=df['Regionname'])

plt.title('jarak to CBD')
plt.ylabel('Region Name')

plt.show()
```

```python
sns.barplot(x=df.Type, y=df.Price, estimator=np.median)

plt.title(('Perbandingan Nilai median harga rumah'), size= 15)
```

```python
sns.scatterplot(x=df.Price, y=df.Distance, hue=df.Type)

plt.title(('Kolerasi Harga Rumah dan jarak rumah ke CBD'), size= 15)
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Month'] = pd.DatetimeIndex(df['Date']).month

plt.figure(figsize=(8,6))
sns.lineplot(x=df.Month, y=df.Price, hue=df.Type)

plt.title(('Tren harga rumah per bulan berdasarkan tipe rumah'), size= 15)
```
